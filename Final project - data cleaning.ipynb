# Final project - data cleaning and preparation for modelling

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

#### Importing the data

```python
data = pd.read_csv("data - data.csv")
```

### Data cleaning

#### I will delete columns that I will not use or are unnecessary.

```python
# Everything unnecessary.

droplist = [i for i in data.columns if "I" in i and "Q" in i or "VCL" in i or "Q" in i and "E" in i]
data.drop(droplist,axis=1,inplace=True)
```

```python
# Everything I will not use. 
# Dropping stress questions.

data.drop(columns = ["screensize", "uniquenetworklocation", "major", "married", "country", "source", "introelapse", "testelapse", "surveyelapse"], inplace = True)
data.drop(columns = ["Q1A", "Q6A", "Q8A", "Q11A", "Q12A", "Q14A", "Q18A", "Q22A", "Q27A", "Q29A", "Q32A", "Q33A", "Q35A", "Q39A"], inplace = True)
```

#### I will check the unique values of the columns to see if anything stands out or needs to be changed.

```python
data["familysize"].value_counts()
```

```python
# There are some too big values for age (like 1998 or 223) and familysize (like 133 or 62).
# As there are not so many of them, I will delete corresponding rows. 

data = data.loc[(data["age"] < 90)]
data = data.loc[(data["familysize"] < 21)]
```

```python
# I will make an age group column (I will only use it in EDA).

bins = [13, 18, 30, 50, 70, 100]
labels = ["-17", "18-29","30-49", "50-69", "70+"]
data["age_group"] = pd.cut(data.age, bins, labels = labels, include_lowest = True)
```

#### I will calculate the summarized scores for depression and anxiety and store them in a new column.

```python
depression_columns = ["Q3A", "Q5A", "Q10A", "Q13A", "Q16A", "Q17A", "Q21A", "Q24A", "Q26A", "Q31A", "Q34A", "Q37A", "Q38A", "Q42A"]
data["depression_score"] = data[depression_columns].sum(axis = 1)
```

```python
anxiety_columns = ["Q2A", "Q4A", "Q7A", "Q9A", "Q15A", "Q19A", "Q20A", "Q23A", "Q25A", "Q28A", "Q30A", "Q36A", "Q40A", "Q41A"]
data["anxiety_score"] = data[anxiety_columns].sum(axis = 1)
```

#### I will add columns for the severeness class for depression and for anxiety - 1 (moderate), 2 (severe), 3 (extremely severe)

```python
conditions_depression = [(data["depression_score"] > 27), (data["depression_score"] > 20), (data["depression_score"] > 13)]
values_depression = [3, 2, 1]
```

```python
data["severeness_level_depression"] = np.select(conditions_depression, values_depression)
```

```python
conditions_anxiety = [(data["anxiety_score"] > 19), (data["anxiety_score"] > 14), (data["anxiety_score"] > 9)]
values_anxiety = [3, 2, 1]
```

```python
data["severeness_level_anxiety"] = np.select(conditions_anxiety, values_anxiety)
```

```python
data["severeness_level_anxiety"].value_counts()
```

#### New column for personality score

```python
# To get the personality scores, I need to calculate together corresponding column values.
# Some have a reverse score, which need to be calculated first to get the real values.

reverse_statements =["TIPI2", "TIPI4", "TIPI6", "TIPI8", "TIPI10"]
reverse_values = {0:0, 1:7, 2:6, 3:5, 4:4, 5:3, 6:2, 7:1}
```

```python
for x in reverse_statements:
    data[x] = data[x].map(reverse_values)
```

```python
# I will calculate the scores for five personality traits.

data["extraversion"] = (data["TIPI1"] + data["TIPI6"])/2
data["agreeableness"] = (data["TIPI2"] + data["TIPI7"])/2
data["conscientiousness"] = (data["TIPI3"] + data["TIPI8"])/2
data["emotional_stability"] = (data["TIPI4"] + data["TIPI9"])/2
data["openness_to_experiences"] = (data["TIPI5"] + data["TIPI10"])/2
```

```python
# I will delete all the TIPI separate statements from the table and only leave the calculated scores.

data.drop(columns = ["TIPI1", "TIPI2", "TIPI3", "TIPI4", "TIPI5", "TIPI6", "TIPI7", "TIPI8", "TIPI9", "TIPI10"], inplace = True)
```

```python
# Correlations between the columns.
# data.corr()

corr = data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 30))
    ax = sns.heatmap(corr, mask=mask,cmap='coolwarm', vmin=-1,vmax=1,annot=True, square=True)
```
